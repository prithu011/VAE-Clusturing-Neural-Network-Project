# Extracting Song ID, Quadrant, and Genre from a CSV Dataset

This guide explains how to process a **CSV dataset containing multiple features**, and extract only three specific columns: **song ID, quadrant information, and genre**. The resulting dataset can then be used for machine learning tasks, such as clustering, classification, or conditional VAE models.


```python
# Extracting Genres
import pandas as pd

input_csv = r"E:\CSE425_Project\project\data\MERGE_Lyrics_Complete\MERGE_Lyrics_Complete\merge_lyrics_complete_metadata.csv"
df = pd.read_csv(input_csv)

print("Original columns:", df.columns)

columns_to_keep = ["Song", "Quadrant", "Title", "Genres"]  

new_df = df[columns_to_keep]

output_csv = r"E:\CSE425_Project\project\data\song_quadrant_genre.csv"
new_df.to_csv(output_csv, index=False)

print(f"New CSV saved: {output_csv}")
```
